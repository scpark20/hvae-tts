```python
import json
import numpy as np
import matplotlib.pyplot as plt
```

```python
dirs = ['save/hvae_lj_parallel18_1/data.json',
        'save/hvae_lj_parallel18_2/data.json',
        'save/hvae_lj_parallel18_3/data.json',
        'save/hvae_lj_parallel18_4/data.json',
       ]

N = 100000

loss_list = []
for dir in dirs:
    with open(dir, 'r') as f:
        loss = np.array(json.load(f)['recon_loss'])
        loss[:, 1] = np.clip(loss[:, 1], a_min=0, a_max=2.0)
        loss_list.append(loss)
    
def moving_average(data_set, periods=100):
    weights = np.ones(periods) / periods
    data_set = np.pad(data_set, (periods-1, 0), mode='edge')
    return np.convolve(data_set, weights, mode='valid')

plt.figure(figsize=[18, 3])
for dir, loss in zip(dirs, loss_list):
    plt.plot(loss[:, 0], moving_average(loss[:, 1]), label=dir, alpha=0.5)
    
plt.xlim([0, N])
plt.ylim([0.004, 0.006])
plt.grid()
plt.legend()
plt.show()

loss_list = []
for dir in dirs:
    with open(dir, 'r') as f:
        loss = np.array(json.load(f)['kl_loss'])
        loss[:, 1] = np.clip(loss[:, 1], a_min=0, a_max=2.0)
        loss_list.append(loss)
    
plt.figure(figsize=[18, 3])
for dir, loss in zip(dirs, loss_list):
    plt.plot(loss[:, 0], moving_average(loss[:, 1]), label=dir, alpha=0.5)
    
plt.xlim([0, N])
plt.ylim([0.0, 0.02])
plt.grid()
plt.legend()
plt.show()

loss_list = []
for dir in dirs:
    with open(dir, 'r') as f:
        loss = np.array(json.load(f)['beta'])
        loss[:, 1] = np.clip(loss[:, 1], a_min=0, a_max=2.0)
        loss_list.append(loss)
    
plt.figure(figsize=[18, 3])
for dir, loss in zip(dirs, loss_list):
    plt.plot(loss[:, 0], loss[:, 1], label=dir, alpha=0.5)
    
plt.xlim([0, N])
plt.ylim([0.0, 1.1])
plt.grid()
plt.legend()
plt.show()
```

```python
def annealing(steps, num_epochs, duration):
    if steps < 0:
        return 0.0
    if steps >= duration:
        return 1.0
    
    steps_per_epoch = duration // num_epochs
    mod_steps = steps % steps_per_epoch
    epochs = steps // steps_per_epoch
    coeff = (epochs + 1) / num_epochs
    value = coeff * (1 - (np.cos(2 * np.pi * mod_steps / steps_per_epoch * 0.5) + 1) / 2)
    
    return value

duration = 100000
steps = []
for step in range(100000):
    steps.append(annealing(step, 100, duration))
    
plt.figure(figsize=[18, 3])
plt.plot(steps)
plt.show()
        
```
